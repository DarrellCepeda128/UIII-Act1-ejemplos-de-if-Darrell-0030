### Ejemplos de if, if else, if elif

### Ejemplo de if

```python
# 1. Ejemplo con if (Uso de operadores aritméticos y relacionales)
# Este ejemplo verifica si el costo total de una compra supera un límite para aplicar un mensaje de agradecimiento.
precio_unitario = float(input("Introduce el precio del producto: "))
cantidad = int(input("¿Cuántos artículos llevas?: "))
limite_promocion = 100

total = precio_unitario * cantidad
print(f"Total a pagar: ${total}")

if total > limite_promocion:
    print("¡Gracias por tu gran compra! Has recibido un mensaje especial.")
```

```python
# 2. Ejemplo con if-else (Uso de operadores lógicos)
# Aquí validamos si una persona puede acceder a un evento basándose en dos condiciones.
edad = int(input("Ingresa tu edad: "))
respuesta = input("¿Tienes entrada? (si/no): ").lower().strip()

tiene_entrada = (respuesta == "si")

if edad >= 18 and tiene_entrada:
    print("Acceso permitido. ¡Disfruta el evento!")
else:
    print("No puedes ingresar. Requisitos: ser mayor de edad y tener entrada.")
```

```python
# 3. Ejemplo con if-elif-else (Combinación de operadores)
# Este código clasifica el rendimiento de un estudiante basado en su puntaje.
puntaje_examen = int(input("Introduce tu puntaje (0-100): "))
asistencia_sn = input("¿Tuviste asistencia completa? (si/no): ").lower().strip()

asistencia_completa = (asistencia_sn == "si")

if puntaje_examen >= 90 or (puntaje_examen >= 80 and asistencia_completa):
    print("Calificación: Excelente ⭐")
elif puntaje_examen >= 60:
    print("Calificación: Aprobado 👍")
else:
    print("Calificación: Reprobado 👎")
```
